```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
datasets_dir = r'..\..\python-ml-course-master\datasets'
filepath = datasets_dir + os.sep + r'customer-churn-model\Customer Churn Model.txt'
```

```python
data = pd.read_csv(filepath)
rows, cols = data.shape
```

```python
figure, axs = plt.subplots(2,2, sharey = True, sharex = True)
data.plot(kind="scatter", x="Day Mins", y="Day Charge", ax=axs[0][0])
data.plot(kind="scatter", x="Night Mins", y="Night Charge", ax=axs[0][1])
data.plot(kind="scatter", x="Day Calls", y="Day Charge", ax=axs[1][0])
data.plot(kind="scatter", x="Night Calls", y="Night Charge", ax=axs[1][1])
```

# Histograma de frecuencias

```python
#plt.hist(data["Day Calls"], bins=20)
M = int(np.ceil(1 + np.log2(rows)))
plt.hist(data["Day Calls"], bins= M) # Regla de Sturges para determinar las divisiones de un histograma
plt.xlabel("NÃºmero de llamadas al dÃ­a")
plt.ylabel("Frecuencia")
plt.title("Histograma del nÃºmero de llamadas al dÃ­a")
```

# Box plot, diagrama de bigotes

```python
plt.boxplot(data["Day Calls"])
plt.ylabel("NÃºmero de llamadas diarias")
plt.title("Boxplot de las llamadas diarias")
```

La lÃ­nea roja representa la mediana o el percentil 50. La lÃ­nea inferior el percentil 25 y el superior el percentil 75. Esos eprcentiles se muestran mÃ¡s arriba con la funciÃ³n "describe".

Rango intercuantilico IQR = valor75% - valor25%.
Los bigotes se suelen poner a 1,5 * IQR de distancia

```python
description = data["Day Calls"].describe()
IQR = description["75%"] - description["25%"]
print('IQR es %s' % IQR)
```

```python
IQR = data["Day Calls"].quantile(0.75) - data["Day Calls"].quantile(0.25)
```

```python
upper_limit = data["Day Calls"].quantile(0.75) + IQR * 1.5
lower_limit = data["Day Calls"].quantile(0.25) - IQR * 1.5
print("LÃ­mite superior %s\nLimite inferior %s" % (upper_limit, lower_limit))
```
